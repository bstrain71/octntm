```python
# Count the types of changes made to the ATONs

import re

# Create a dictionary for counting
atonCounts = dict()
atonCounts["ATON Changes"] = 0
atonCounts["ATON Established"] = 0
atonCounts["ATON Discontinued"] = 0

for idx, row in ntm.iterrows():
  #print(idx)
  # Open the file in read mode
  if ntm['descriptor'].iat[idx] == "ATON":
    text = ntm['body'].iat[idx]

    # The header of the section I want to search.
    atonChanges = "THE FOLLOWING AID TO NAVIGATION CHANGES HAVE BEEN MADE"
    if re.search(atonChanges, text):
      startChanges = re.search(atonChanges, text).span()[1] # The index of where the header ends - upper bound of the search.
      endChanges = re.search(r"THE FOLLOWING|CANCEL", text[startChanges:]).span()[1] + startChanges # Index of where the next section begins (following or cancel). This index PLUS the upper bound is the lower bound. re.search gives the relative index not the absolute index.
      countChanges = re.findall(r"\n[^\W\d_]\.", text[startChanges:endChanges]) # List of headers to change bullets. The length of this list is the number of changes.\
      countChanges = len(countChanges)

    atonEstablished = "THE FOLLOWING AIDS TO NAVIGATION HAVE BEEN ESTABLISHED"
    if re.search(atonEstablished, text):
      startEstablished = re.search(atonEstablished, text).span()[1]
      endEstablished = re.search(r"THE FOLLOWING|CANCEL", text[startEstablished:]).span()[1] + startEstablished
      countEstablished = re.findall(r"\n[^\W\d_]\.", text[startEstablished:endEstablished])
      countEstablished = len(countEstablished)

    atonDiscontinued = "THE FOLLOWING AIDS TO NAVIGATION HAVE BEEN DISCONTINUED"
    if re.search(atonDiscontinued, text):
      startDiscontinued = re.search(atonDiscontinued, text).span()[1]
      endDiscontinued = re.search(r"THE FOLLOWING|CANCEL", text[startDiscontinued:]).span()[1] + startDiscontinued
      countDiscontinued = re.findall(r"\n[^\W\d_]\.", text[startDiscontinued:endDiscontinued])
      countDiscontinued = len(countDiscontinued)

    atonCounts["ATON Changes"] = atonCounts["ATON Changes"] + countChanges
    atonCounts["ATON Established"] = atonCounts["ATON Established"] + countEstablished
    atonCounts["ATON Discontinued"] = atonCounts["ATON Discontinued"] + countDiscontinued

atonCounts
```

```python
# Setupt to make the powerpoint
from pptx import Presentation
from pptx.util import Inches
```

```python
# Create a blank presentation
prs = Presentation()
```

```python
# Title Slide
bullet_slide_layout = prs.slide_layouts[1]

titleSlide = prs.slides.add_slide(bullet_slide_layout)
titleShapes = titleSlide.shapes

title_shape = titleShapes.title
body_shape = titleShapes.placeholders[1]

title_shape.text = f"Automatically Generated PPTx Demo"

tf = body_shape.text_frame
tf.text = f"This ppt was generated automatically using raw text from USCG Broadcast NTM from \n {min(ntm['date'])} to {max(ntm['date'])}"

p = tf.add_paragraph()
p.text = 'Code Repository: https://github.com/bstrain71/octntm'
p.level = 1
```

```python
# Messages Sent Slide
title_only_slide_layout = prs.slide_layouts[5]
slide = prs.slides.add_slide(title_only_slide_layout)
shapes = slide.shapes

shapes.title.text = f"\nUSCG {ntm['originator'].iat[0]} Messages Summary"

tableMsgs = shapes.add_table(6, #rows
                             2, #cols
                             Inches(2.0), # left
                             Inches(2.0), # top
                             Inches(6.0), # width
                             Inches(0.8) # height
                             ).table

# set column widths
tableMsgs.columns[0].width = Inches(2.0)
tableMsgs.columns[1].width = Inches(4.0)

# write column headings
tableMsgs.cell(0, 0).text = 'Type'
tableMsgs.cell(0, 1).text = 'Number Sent'

# write body cells
tableMsgs.cell(1, 0).text = descriptors[0]
tableMsgs.cell(1, 1).text = str(ntm[ntm.descriptor == descriptors[0]].shape[0])
tableMsgs.cell(2, 0).text = descriptors[1]
tableMsgs.cell(2, 1).text = str(ntm[ntm.descriptor == descriptors[1]].shape[0])
tableMsgs.cell(3, 0).text = descriptors[2]
tableMsgs.cell(3, 1).text = str(ntm[ntm.descriptor == descriptors[2]].shape[0])
tableMsgs.cell(4, 0).text = descriptors[3]
tableMsgs.cell(4, 1).text = str(ntm[ntm.descriptor == descriptors[3]].shape[0])
tableMsgs.cell(5, 0).text = descriptors[4]
tableMsgs.cell(5, 1).text = str(ntm[ntm.descriptor == descriptors[4]].shape[0])
```
